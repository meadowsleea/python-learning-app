## Lesson 1: Introduction to Pandas
`pandas` is a powerful library for data manipulation and analysis.

Let's start by importing it and loading a sample dataset.

```
import pandas as pd
import matplotlib.pyplot as plt

# Load a sample dataset
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
df = pd.read_csv(url)
df.head()
```

## Lesson 2: Data Cleaning
Cleaning data involves handling missing values, renaming columns, and converting data types.

```
# Check for missing values
df.isnull().sum()

# Rename columns
df.rename(columns={'total_bill': 'TotalBill', 'tip': 'Tip'}, inplace=True)
df.head()
```

## Lesson 3: Data Filtering
You can filter data using conditions.

```
# Filter rows where Tip is greater than 5
df[df['Tip'] > 5]
```

## Lesson 4: Data Visualization
Use `matplotlib` to create plots and visualize data.

```
# Create a scatter plot of TotalBill vs Tip
plt.figure(figsize=(8,6))
plt.scatter(df['TotalBill'], df['Tip'], alpha=0.6)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs Tip')
plt.grid(True)
plt.show()
```
